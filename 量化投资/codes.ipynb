```python
import pandas as pd
import os
import datetime
index_1 = 0
index_2 = 0
df_list = pd.read_excel('/Users/liuhuanshuo/Desktop/作业/量化/stocks_list.xlsx')
```

```python
df = pd.DataFrame(columns = ['沪A股票代码','日期','市盈率'])
```

```python
#选出20160401日股票
#path = '/Users/liuhuanshuo/desktop/作业/单佳雷数据/沪A股票原始数据/'
for file in df_list['股票代码']:
    excel_path = os.path.join('/Users','liuhuanshuo','desktop','作业','量化','单佳雷数据','沪A股票原始数据/') + file + '.xls'
    df_one = pd.read_excel(excel_path)
    df_one = df_one.loc[:,['代码','日期','市盈率']]
    num_row = len(df_one['代码'])
    for i in range(0,num_row):
        if df_one.iloc[i,1] == datetime.datetime(2016,4,1) and df_one.iloc[i,2] > 0:
            df.loc[index_1] = [df_one.iloc[i,0],df_one.iloc[i,1],df_one.iloc[i,2]]
            index_1 += 1
            break
```

```python
#市盈率排序
df_sort = df.sort_values(by='市盈率')
index_2 = 0
df_50 = pd.DataFrame(columns = ['沪A股票代码','日期','市盈率'])
for j in range(50):
    df_50.loc[index_2] = [df_sort.iloc[j,0],df_sort.iloc[j,1],df_sort.iloc[j,2]]
    index_2 += 1
    
df_50
```

```python
df_date = pd.read_excel('/Users/liuhuanshuo/Desktop/作业/量化/20160401至20160601的日期.xlsx')
df_alldaysprofit = pd.DataFrame(columns = ['日期','代码','买入日期','买入价','当日收盘价','该股票当日盈亏'])
```

```python
flag = 0
index_3 = 0
```

```python
for file in df_50['沪A股票代码']:
    excel_path = os.path.join('/Users','liuhuanshuo','desktop','作业','量化','单佳雷数据','沪A股票原始数据/') + file + '.xls'
    df_oneof50 = pd.read_excel(excel_path)
    df_oneof50 = df_oneof50.loc[:,['代码','日期','开盘价(元)','收盘价(元)']]
    buy_date = datetime.datetime(2016,4,1)
    buy_price = df_oneof50[df_oneof50['日期'] == buy_date].iloc[0,2]
    
    for i in range(len(df_date)):
        for j in range(len(df_oneof50)):
            if df_oneof50.iloc[j,1] == df_date.iloc[i,0]:
                date = df_date.iloc[i,0]
                close_price = df_oneof50.iloc[j,3]
                profit = 100 * (close_price - buy_price)
                df_alldaysprofit.loc[index_3] = [date, df_oneof50.iloc[j,0],buy_date,buy_price,close_price,profit]
                index_3 += 1
                break
df_alldaysprofit
```

```python
index_4 = 0
index_5 = 0
count = 0 #持仓股票个数
profit_today = 0 #当日组合盈亏
profit_accumulate = 0 #组合累计盈亏
```

```python
df_last = pd.DataFrame(columns = ['日期','当日持仓股票个数','当日组合盈亏','当日组合累计收益'])
df_ProfitOfOneDay = pd.DataFrame(columns = ['日期','当日持仓股票个数','当日组合盈亏'])
```

```python
for i in range(len(df_date)):
    for j in range(len(df_alldaysprofit)):
        if df_date.iloc[i,0] == df_alldaysprofit.iloc[j,0]:
            count += 1
            profit_today = profit_today + df_alldaysprofit.iloc[j,5]
    df_ProfitOfOneDay.loc[index_4] = [df_date.iloc[i,0], count, profit_today]
    index_4 += 1
    count = 0
```

```python
profit_accumulate = df_ProfitOfOneDay.iloc[0,2]
df_last.loc[index_5] = [df_ProfitOfOneDay.iloc[0,0],df_ProfitOfOneDay.iloc[0,1],df_ProfitOfOneDay.iloc[0,2],profit_accumulate]
index_5 += 1
```

```python
for k in range(1,len(df_ProfitOfOneDay)):
    profit_accumulate = profit_accumulate + df_ProfitOfOneDay.iloc[k,2]
    df_last.loc[index_5] = [df_ProfitOfOneDay.iloc[k,0],df_ProfitOfOneDay.iloc[k,1],df_ProfitOfOneDay.iloc[k,2],profit_accumulate]
    index_5 += 1
df_last
```
