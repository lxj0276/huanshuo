```python
import pandas as pd
import os
```

```python
df_list = pd.read_excel('/Users/liuhuanshuo/Desktop/作业/量化/stocks_list.xlsx')
temp = pd.DataFrame(columns = ['代码','日期','市盈率']) #创建用于储存的空数据框
for file in df_list['股票代码']:
    excel_path = os.path.join('/Users','liuhuanshuo','desktop','作业','量化','单佳雷数据','沪A股票原始数据/') + file + '.xls'
    df = pd.read_excel(excel_path) 
    df_one = df.loc[:,['代码','日期','市盈率']]  #切片 提取指定列
    a = df_one[(df_one.日期 == '2016/4/1') & (df_one.市盈率 > 0)] #选择特定行
    temp = temp.append(a,ignore_index=True) #将刚刚选取的行添加到数据框中，可以帮助忽略index，自动递增
```

```python
temp_sort = temp.sort_values(by = '市盈率') #排序
s = temp_sort.iloc[0:50] #选择前50行
s = s.reset_index(drop=True) #用reset_index去处理index,不然容易出现逻辑错误
s
```

```python
df_date = pd.read_excel('/Users/liuhuanshuo/Desktop/作业/量化/20160401至20160601的日期.xlsx')
b = pd.DataFrame(columns = ['代码','日期','收盘价(元)'])
df_alldaysprofit = pd.DataFrame(columns = ['日期','代码','买入日期','买入价','收盘价(元)'])
for file in s['代码']:
    excel_path = os.path.join('/Users','liuhuanshuo','desktop','作业','量化','单佳雷数据','沪A股票原始数据/') + file + '.xls'
    df_oneof50 = pd.read_excel(excel_path)
    for i in df_date.日期:
        i = i.strftime('%Y/%m/%d')
        t = df_oneof50[df_oneof50['日期'] == i]
        b = b.append(t,ignore_index=True)
        b = b[['日期','代码','收盘价(元)']]
        b.insert(2,'买入日期','2016-04-01')
        b.insert(3,'买入价','4.5792')
df_alldaysprofit = df_alldaysprofit.append(b)
```
